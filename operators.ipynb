```python
'''
PEMDAS
P  → Parentheses () 1ST PRORITY
E  → Exponent **2ND PRORITY
M  → Multiply * OR % 3RD PRORITY
D  → Divide /
A  → Add + 4TH PRORITY
S  → Subtract -
'''


print(3*3+3/3-3) #PEMDAS
print(3*(3+3)/3-3) #PEMDAS
print(3*3+3/3-3) #PEMDAS
```

```python
'''
ARITHMETIC OPERATORS
'''
x = 10
y = 20  

print(f"Addition : {x + y}")
print(f"Sub : {x - y}")
print(f"Mul : {x * y}")
print(f"Division : {x / y}")
print(f"Modulus : {x % y}")#modulus % - remainder 
print(f"Exponent : {x ** y}")#power 
print(f"Floor division : {x // y}")#floor division - remove decimal
```

```python
# Variables
x = 10       
y = 5        

# Comparison Operators

print(x == y)   # Equal to: check karta hai x aur y same hain ya nahi
print(x != y)   # Not equal to: check karta hai x aur y different hain ya nahi
print(x > y)    # Greater than: x y se bada hai ya nahi
print(x < y)    # Less than: x y se chhota hai ya nahi
print(x >= y)   # Greater than or equal to: x y se bada ya barabar hai
print(x <= y)   # Less than or equal to: x y se chhota ya barabar hai
```

```python
'''
PEMDAS
P  → Parentheses () 1ST PRORITY
E  → Exponent **2ND PRORITY
M  → Multiply * OR % 3RD PRORITY
D  → Divide /
A  → Add + 4TH PRORITY
S  → Subtract -
'''


print(3*3+3/3-3) #PEMDAS
print(3*(3+3)/3-3) #PEMDAS
print(3*3+3/3-3) #PEMDAS
```

```python
'''
ARITHMETIC OPERATORS
'''
x = 10
y = 20  

print(f"Addition : {x + y}")
print(f"Sub : {x - y}")
print(f"Mul : {x * y}")
print(f"Division : {x / y}")
print(f"Modulus : {x % y}")#modulus % - remainder 
print(f"Exponent : {x ** y}")#power 
print(f"Floor division : {x // y}")#floor division - remove decimal
```

```python
# Variables
x = 10       
y = 5        

# Comparison Operators

print(x == y)   # Equal to: check karta hai x aur y same hain ya nahi
print(x != y)   # Not equal to: check karta hai x aur y different hain ya nahi
print(x > y)    # Greater than: x y se bada hai ya nahi
print(x < y)    # Less than: x y se chhota hai ya nahi
print(x >= y)   # Greater than or equal to: x y se bada ya barabar hai
print(x <= y)   # Less than or equal to: x y se chhota ya barabar hai
```

```python
# Logical operators multiple conditions ko combine karne ke liye use hote hain
# Result hamesha True ya False hota hai


a = 10
b = 20
print(a>5 and b>10) # and → dono True hone chahiye
print(a<1 or b == 20) # or  → koi ek True ho
print(not(a>5 and b>10)) #not operator - reverse the result
```

```python
# =========================
# 4️⃣ ASSIGNMENT OPERATORS
# =========================

num = 10

num += 5        # num = num + 5
print(num)

num -= 3        # num = num - 3
print(num)

num *= 2        # num = num * 2
print(num)

num /= 4        # num = num / 4
print(num)

num %= 3        # num = num % 3
print(num)

num //= 2       # num = num // 2
print(num)

num **= 3       # num = num ** 3
print(num)

```

```python
# =========================
# 5️⃣ IDENTITY OPERATORS
# =========================

x = 10
y = 10
z = 20

print(x is y)       # is → same memory location
print(x is not z)   # is not → different memory location
```

```python
# =========================
# 6️⃣ MEMBERSHIP OPERATORS
# =========================

data = [10, 20, 30, 40]

print(10 in data)        # in → element present hai ya nahi
print(50 not in data)   # not in → element present nahi hai
```

```python
# =========================
# 7️⃣ BITWISE OPERATORS
# (Basic awareness – beginner level)
# =========================

a = 5   # 0101
b = 3   # 0011

print(a & b)   # AND
print(a | b)   # OR
print(a ^ b)   # XOR
print(~a)      # NOT
print(a << 1)  # Left shift
print(a >> 1)  # Right shift
```

```python
'''
Python supports 2 types of iterative statements. 
 
1. for loop -
        Syntax: 
 
    for x in sequence : 
    #code
2 . while loop
'''
name="Khushi Navik"   
for x in name :   
    print(x)  
```

```python
'''
range()
parameters:start=1, stop=100, and step=1
'''

nums = tuple(range(1,100,9))
print(nums)
```

```python
'''
Nested Loops: 
for outer_variable in range(...):
    for inner_variable in range(...):
        # code to execute

'''
for i in range(1,3):
    for j in range(3,6):
        print(f'{i},{j}')
```

```python
for i in range(1, 6):          # table number
    print(f"\nTable of {i}")
    for j in range(1, 11):     # multiplier
        print(f"{i} x {j} = {i * j}")
```

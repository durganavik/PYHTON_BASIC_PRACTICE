```python
def greet(name):#name- parameter
    print(f"Good Morning, {name}")
    print("Hello",name)


greet("Khushi")#khushi - agrument
```

```python
'''
types of arguments=>positional,keyword,default
'''
def greet(city,name):# positional parameter
    print(f"Wel {name} to our {city}")
   


greet("Khushi","Surat")
```

```python
def greet(name="khushi",city="surat"):# default parameter
    print(f"Wel {name} to our {city}")
   


greet()
```
